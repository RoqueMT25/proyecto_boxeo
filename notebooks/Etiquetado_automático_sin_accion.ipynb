```python
import os
import cv2
import numpy as np
import re
import csv
```

```python
# Ruta a la carpeta de fotogramas
carpeta = r"C:\Users\Usuario\Desktop\mi_proyecto\1sin_accion"

# Obtener la lista de archivos en la carpeta
archivos = os.listdir(carpeta)
archivos = [archivo for archivo in archivos if archivo.endswith('.jpg') or archivo.endswith('.jpeg') or archivo.endswith('.png')]

# Función para extraer el último número del nombre del archivo
def extraer_ultimo_numero(nombre):
    numeros = re.findall(r'\d+', nombre)
    if numeros:
        return numeros[-1]  # Usar el último número encontrado
    return None

# Crear una lista para almacenar los nuevos nombres de los archivos
archivos_renombrados = []

# Renombrar los archivos solo con el último número encontrado en el nombre
for archivo in archivos:
    ultimo_numero = extraer_ultimo_numero(archivo)
    if ultimo_numero:
        base_nombre = ultimo_numero
        extension = os.path.splitext(archivo)[1]  # Obtener la extensión del archivo
        nuevo_nombre = base_nombre + extension
        ruta_antigua = os.path.join(carpeta, archivo)
        ruta_nueva = os.path.join(carpeta, nuevo_nombre)
        try:
            os.rename(ruta_antigua, ruta_nueva)
            archivos_renombrados.append(nuevo_nombre)
            print(f"Renombrando {archivo} a {nuevo_nombre}")
        except Exception as e:
            print(f"No se pudo renombrar {archivo}: {e}")

print("Renombrado completado.")

# Ruta y nombre del archivo CSV
csv_path = r"C:\Users\Usuario\Desktop\mi_proyecto\archivos_lista.csv"

# Escribir los nombres de los archivos renombrados en el archivo CSV
with open(csv_path, mode='w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['Nombre de Archivo'])  # Escribir el encabezado
    for archivo in archivos_renombrados:
        csv_writer.writerow([archivo])

print(f"Archivo CSV creado en {csv_path}")
```

```python
# Ruta a la carpeta de fotogramas
carpeta = r"C:\Users\Usuario\Desktop\mi_proyecto\1sin_accion"

# Ruta y nombre del archivo CSV
csv_path = r"C:\Users\Usuario\Desktop\mi_proyecto\archivos_lista.csv"

# Leer los nombres de los archivos desde el archivo CSV
with open(csv_path, mode='r') as csvfile:
    csv_reader = csv.reader(csvfile)
    next(csv_reader)  # Saltar el encabezado
    archivos = [row[0] for row in csv_reader]

# Ordenar los archivos por su nombre numérico
archivos.sort(key=lambda x: int(os.path.splitext(x)[0]))

# Función para generar el nuevo nombre del archivo según el patrón especificado
def generar_nuevo_nombre(indice_secuencia, indice_archivo):
    return f"sin_accion_{indice_secuencia:03d}_{indice_archivo:02d}.jpg"

# Renombrar los archivos según el patrón especificado
indice_secuencia = 1
indice_archivo = 1
ultimo_numero = int(os.path.splitext(archivos[0])[0])

for archivo in archivos:
    ruta_antigua = os.path.join(carpeta, archivo)
    if not os.path.exists(ruta_antigua):
        print(f"Archivo no encontrado: {ruta_antigua}")
        continue

    numero_actual = int(os.path.splitext(archivo)[0])
    
    if numero_actual == ultimo_numero + 1:
        nuevo_nombre = generar_nuevo_nombre(indice_secuencia, indice_archivo)
        indice_archivo += 1
    else:
        indice_secuencia += 1
        indice_archivo = 1
        nuevo_nombre = generar_nuevo_nombre(indice_secuencia, indice_archivo)
        indice_archivo += 1
    
    ultimo_numero = numero_actual
    
    ruta_nueva = os.path.join(carpeta, nuevo_nombre)
    
    try:
        os.rename(ruta_antigua, ruta_nueva)
        print(f"Renombrando {archivo} a {nuevo_nombre}")
    except Exception as e:
        print(f"No se pudo renombrar {archivo}: {e}")

print("Renombrado completado.")
```
