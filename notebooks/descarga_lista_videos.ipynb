```python
import subprocess
import os
```

```python
def download_youtube_video(url, n):
    # Aseg√∫rate de que la carpeta videos existe
    if not os.path.exists(n):
        os.makedirs(n)

    # Define el comando yt-dlp con los argumentos necesarios
    command = [
        'yt-dlp', 
        '-f', 'best', 
        '-o', f'{n}/%(title)s.%(ext)s', 
        url
    ]
    
    try:
        # Ejecuta el comando y captura la salida y errores
        result = subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
        print("Video descargado exitosamente.")
        print("Salida del comando:", result.stdout)
    except subprocess.CalledProcessError as e:
        print(f"Error al descargar el video: {e}")
        print("Error detallado:", e.stderr)
```
